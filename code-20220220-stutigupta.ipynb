```python
#Importing libraries
import json
import pandas as pd
```

```python
#Reading data from json file
with open('sample.json', 'r') as openfile:
  
    # Reading from json file
    json_object = json.load(openfile)
    
print(json_object)
print(type(json_object))
```

```python
#Converting json_object to dataframe
df = pd.DataFrame(json_object)
```

```python
df.head(5)
```

```python
#Changing height from cm to meters and squaring it
df['HeightM2'] = (df['HeightCm']/100)**2
```

```python
#Calculating BMI
df['BMI'] = df['WeightKg']/df['HeightM2']
```

```python
df.head()
```

```python
#Bringing the BMI category information
stand = {
            'BMI_category': ['Underweight','Normal','Overweight','Moderately','Severely','Very'],
            'BMI_Range_low_limit':[0, 18.5, 25, 30, 35, 40],
            'BMI_Range_high_limit':[18.4, 24.9, 29.9, 34.9, 39.9,1000000000],
            'Health_Risk' : ['Malnutrition risk', 'Low risk', 'Enhanced risk', 'Medium risk', 'High risk','Very high risk']
            
}


df_stand = pd.DataFrame(stand) # Changing the dictionary to dataframe
```

```python
df_stand.head(5)
```

```python
df1 = pd.merge(df, df_stand,how='cross') #Cross joining both the tables to find BMI_category and Health_risk
```

```python
df1.head(10)
```

```python
#Filtering rows as per the range values
df2 = df1[(df1.BMI >= df1.BMI_Range_low_limit) & (df1.BMI <= df1.BMI_Range_high_limit)] 
```

```python
df2.head(10)
```

```python
#Dropping extra columns
df3 = df2.drop(['BMI_Range_low_limit','BMI_Range_high_limit'], axis =1)
```

```python
df3.head()
```

```python
#One person is overweight
df4 = df3.groupby(['BMI_category']).size()
df4.head()
```
